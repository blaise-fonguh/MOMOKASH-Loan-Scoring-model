```python
import pandas as pd
import glob
import os
```

```python
# Merge all loan files from loan_data folder
loan_files = glob.glob("loan_data/*.csv")
loan_dfs = [pd.read_csv(f) for f in loan_files]
merged_loans = pd.concat(loan_dfs, ignore_index=True)
merged_loans.to_csv("merged_loans_all.csv", index=False)
print(f"Merged {len(loan_files)} loan files")

```

```python
# Merge all refund files from refund_data folder
refund_files = glob.glob("refund_data/*.csv")
refund_dfs = [pd.read_csv(f) for f in refund_files]
merged_refunds = pd.concat(refund_dfs, ignore_index=True)
merged_refunds.to_csv("merged_refunds_all.csv", index=False)
print(f"Merged {len(refund_files)} refund files")

print("\nâœ… All files merged successfully!")
print(f"Merged {len(loan_files)} loan files")

```

```python
# Merge all client files from client_data folder
client_files = glob.glob("client_data/*.csv")
client_dfs = [pd.read_csv(f) for f in client_files]
merged_clients = pd.concat(client_dfs, ignore_index=True)
merged_clients.to_csv("merged_clients_all.csv", index=False)
print(f"Merged {len(client_files)} client files")
```
